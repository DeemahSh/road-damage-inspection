```python
import sys, random, os, shutil
from os import listdir
from os.path import isfile, join 

def splitdirs(files, dir1, dir2, dir3, dir4, ratio):
    shuffled = files[:]
    random.shuffle(shuffled)
    num = round(len(shuffled) * ratio)
    to_dir1, to_dir2 = shuffled[:num], shuffled[num:]
    for d in dir1, dir2:
        if not os.path.exists(d):
            os.mkdir(d)
    for file in to_dir1:
        shutil.copy2(os.path.join('./RoadDamageDataset/All_Images', os.path.basename(file)), os.path.join(dir1, os.path.basename(file)))
        annotation_file = file.replace('jpg','xml')
        shutil.copy2(os.path.join('./RoadDamageDataset/All_Annotations', os.path.basename(annotation_file)), os.path.join(dir3, os.path.basename(annotation_file)))
    for file in to_dir2:
        shutil.copy2(os.path.join('./RoadDamageDataset/All_Images', os.path.basename(file)), os.path.join(dir2, os.path.basename(file)))
        annotation_file = file.replace('jpg','xml')
        shutil.copy2(os.path.join('./RoadDamageDataset/All_Annotations', os.path.basename(annotation_file)), os.path.join(dir4, os.path.basename(annotation_file)))
```
